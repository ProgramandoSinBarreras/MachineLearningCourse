```python
import os
import shutil
from google.colab import drive

# Monta Google Drive
drive.mount('/content/drive')

# Clona el repositorio de Git
!git clone https://github.com/ProgramandoSinBarreras/MachineLearningCourse.git
```

```python
# Define la ruta del repositorio clonado y la ruta de destino en Drive
ruta_repositorio = '/content/MachineLearningCourse'
ruta_destino_drive = '/content/drive/My Drive/Machine_Learning_Course/'

# Crea la ruta de destino si no existe
os.makedirs(ruta_destino_drive, exist_ok=True)
def copiar_si_no_existe(ruta_origen, ruta_destino):
    # Si el origen es un directorio, copia cada archivo/directorio dentro de él
    if os.path.isdir(ruta_origen):
        os.makedirs(ruta_destino, exist_ok=True)  # Asegura que el directorio de destino existe
        for elemento in os.listdir(ruta_origen):
            sub_ruta_origen = os.path.join(ruta_origen, elemento)
            sub_ruta_destino = os.path.join(ruta_destino, elemento)
            copiar_si_no_existe(sub_ruta_origen, sub_ruta_destino)  # Recursión para copiar contenido
    else:
        # Verifica si el archivo ya existe en el destino
        if not os.path.exists(ruta_destino):
            # Si el archivo no existe, lo copia al destino en Drive
            shutil.copy2(ruta_origen, ruta_destino)
            print(f'Archivo {os.path.basename(ruta_destino)} copiado a Drive.')
        else:
            print(f'Archivo {os.path.basename(ruta_destino)} ya existe en Drive. No se copió para evitar duplicados.')

# Itera sobre los archivos en el directorio del repositorio clonado
for archivo in os.listdir(ruta_repositorio):
    ruta_archivo_origen = os.path.join(ruta_repositorio, archivo)
    ruta_archivo_destino = os.path.join(ruta_destino_drive, archivo)
    copiar_si_no_existe(ruta_archivo_origen, ruta_archivo_destino)
```
