#### 1.2.1 Slicing

Python ofrece una sintaxis concisa para acceder a sublistas, lo que se conoce como slicing. Esta t칠cnica permite extraer una porci칩n de una lista especificando un rango de 칤ndices.

````python
#@title **Sublistas**
nums = list(range(5))     # range is a built-in function that creates a list of integers
print(type(nums))
nums = [0,1,2,3,4]
print(nums)
print(type(nums))         # Prints "[0, 1, 2, 3, 4]"
print(nums[2:4])          # Get a slice from index 2 to 4 (exclusive); prints "[2, 3]"
print(nums[2:])           # Get a slice from index 2 to the end; prints "[2, 3, 4]"
print(nums[:2])           # Get a slice from the start to index 2 (exclusive); prints "[0, 1]"
print(nums[:])            # Get a slice of the whole list; prints "[0, 1, 2, 3, 4]"
print(nums[:-1])          # Slice indices can be negative; prints "[0, 1, 2, 3]"
nums[2:4] = [8, 9]        # Assign a new sublist to a slice
print(nums)
````

### 2.4 Vectorizaci칩n

Vectorizaci칩n se refiere al proceso de optimizar el c칩digo en Python para que se ejecute m치s r치pido, evitando el uso de estructuras de bucles. En lugar de procesar los elementos de un array uno por uno en un bucle, se realizan operaciones sobre arrays completos.

**producto interno**
El "dot product" entre dos vectores $a$ y $b$ se define como:

$$a_{1 \times n} \cdot b_{n \times 1} = c,$$

aqu칤, $a$ es un vector fila de tama침o $1 \times n$, $b$ es un vector columna de tama침o $n \times 1$, y el resultado $c$ es un escalar.

````python
#@title **code** vectorization performance
import time
a = np.random.randint(10, size=10000)
b = np.random.randint(10, size=10000)

#  Classical way
tic = time.process_time()
dot = 0.0;

for i in range(len(a)):
      dot += a[i] * b[i]

toc = time.process_time()
print("dot_product = "+ str(dot));
print("Computation time = " + str(1000*(toc - tic )) + "ms")

#  vectorized way

n_tic = time.process_time()
n_dot_product = np.dot(a, b)
n_toc = time.process_time()

print("\nn_dot_product = "+str(n_dot_product))
print("Computation time = "+str(1000*(n_toc - n_tic ))+"ms")
````

Otra forma de realizar la multiplicaci칩n es mediante el **producto externo** (tambi칠n conocido como producto tensorial), que genera una matriz a partir de dos vectores. Este producto se representa matem치ticamente como:

$$a_{n\times1} \otimes b^{T}_{1\times n} = c_{n\times n},$$

donde:
- $a$ es un vector columna de dimensiones $n \times 1$.
- $b^{T}$ es el vector fila transpuesto de dimensiones $1 \times n$.
- El resultado $c$ es una matriz de dimensiones $n \times n$.

````python
#@title **code**
#@markdown classical way
tic = time.process_time()
outer_product = np.zeros((10000, 10000))

for i in range(len(a)):
    for j in range(len(b)):
        outer_product[i][j]= a[i]*b[j]

toc = time.process_time()
print("outer_product = "+ str(outer_product));
print("Computation time = "+str(1000*(toc - tic ))+"ms")
#@markdown vectorized way
n_tic = time.process_time()
outer_product = np.outer(a, b)
n_toc = time.process_time()

print("outer_product = "+str(outer_product));
print("\nComputation time = "+str(1000*(n_toc - n_tic ))+"ms")
````

La siguiente expresi칩n es com칰n en modelos de predicci칩n lineal, donde la salida $y^i$ es el resultado de la suma de las entradas $x_j^i$ ponderadas por los par치metros $\theta_j^i$. Al escribir la expresi칩n de forma vectorizada, se optimiza la eficiencia computacional, ya que las operaciones sobre matrices y vectores en Python (con bibliotecas como NumPy) son altamente optimizadas.

$$y^i = \sum_{j=0}^{n} \theta_{j}^{i} x_{j}^{i} = \Theta X^\top; \, i=\{1, 2\}.$$

````python
#@title **code**
#@markdown elementwise operations
theta = np.array([0.4,0.04,0.3,0.2,0.5])
X = np.array([[1,5,8,9,7],
              [1,3,4,5,6]])
y = np.array([0.0,0.0])

for m in list(range(2)):
    result = 0
    for n in list(range(5)):
        result = result + (theta[n]*X[m,n])
    y[m] = result

print(y)
#@markdown Dot product
y = np.dot(theta,X.T)
print(y)
````

````python
#@title **code** Radiaci칩n de cuerpo negro
import numpy as np
import matplotlib.pyplot as plt

# Constantes
h = 6.626e-34  # Constante de Planck (J췅s)
c = 3e8  # Velocidad de la luz (m/s)
k_b = 1.38e-23  # Constante de Boltzmann (J/K)

# Funci칩n de Planck para la radiaci칩n de cuerpo negro
def planck(wavelength, T):
    return (2 * h * c**2) / (wavelength**5) * (1 / (np.exp((h * c) / (wavelength * k_b * T)) - 1))

# Longitudes de onda en metros
wavelengths = np.linspace(1e-9, 3e-6, 500)

# Espectros para diferentes temperaturas
T1 = 3000  # Kelvin
T2 = 4000
T3 = 5000

# Graficar los espectros
plt.plot(wavelengths * 1e9, planck(wavelengths, T1), label='T = 3000 K')
plt.plot(wavelengths * 1e9, planck(wavelengths, T2), label='T = 4000 K')
plt.plot(wavelengths * 1e9, planck(wavelengths, T3), label='T = 5000 K')

plt.xlabel('Longitud de onda (nm)')
plt.ylabel('Intensidad')
plt.title('Radiaci칩n de Cuerpo Negro - Ley de Planck')
plt.legend()
plt.grid(True)
plt.show()
````

### 3.2 Subplots

El m칠todo `subplot()` permite mostrar m칰ltiples gr치ficos en la misma figura.

#### 2D

````python
#@title **code** Subplot en 2D
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
````

````python
#@title **code** Tipos de gr치ficas
# Crear datos para el gr치fico de dispersi칩n
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
sizes = 100 * np.random.rand(50)
colors = np.random.rand(50)

# Crear datos para el gr치fico de barras
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [3, 7, 2, 5, 9]

# Crear figura y subplots
fig, axs = plt.subplots(1, 2, figsize=(8, 4))  # 1 fila, 2 columnas

# Gr치fico de dispersi칩n (scatter plot)
axs[0].scatter(x_scatter, y_scatter, s=sizes, c=colors, alpha=0.5)
axs[0].set_title('Gr치fico de Dispersi칩n')
axs[0].set_xlabel('Eje X')
axs[0].set_ylabel('Eje Y')

# Gr치fico de barras (bar plot)
axs[1].bar(categorias, valores, color='blue')
axs[1].set_title('Gr치fico de Barras')
axs[1].set_xlabel('Categor칤as')
axs[1].set_ylabel('Valores')

# Mostrar el gr치fico
plt.tight_layout()
plt.show()
````

#### 3D

````python
#@title **code** Subplot en 3D
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Crear una figura para los tres gr치ficos
fig = plt.figure(figsize=(12, 6))

# ================== Gr치fico de elipsoide ==================
# Par치metros del elipsoide
a = 2  # Semieje en el eje x
b = 1  # Semieje en el eje y
c = 1.5  # Semieje en el eje z

# Crear una malla esf칠rica para el elipsoide
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x_ellipsoid = a * np.outer(np.cos(u), np.sin(v))
y_ellipsoid = b * np.outer(np.sin(u), np.sin(v))
z_ellipsoid = c * np.outer(np.ones_like(u), np.cos(v))

# Primer subplot: Elipsoide
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(x_ellipsoid, y_ellipsoid, z_ellipsoid, color='b', alpha=0.6)
ax1.set_title('Elipsoide 3D')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')
ax1.set_zlabel('Eje Z')

# ================== Gr치fico de l칤neas en 3D ==================
# Crear datos para el gr치fico de l칤neas 3D
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)
z_line = np.cos(x_line)

# Segundo subplot: L칤nea en 3D
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot(x_line, y_line, z_line)
ax2.set_title('Gr치fico de L칤neas en 3D')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')
ax2.set_zlabel('Eje Z')

# ================== Gr치fico de dispersi칩n en 3D ==================
# Crear datos para el gr치fico de dispersi칩n 3D
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
z_scatter = np.random.rand(50)

# Tercer subplot: Dispersi칩n en 3D
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(x_scatter, y_scatter, z_scatter, c='r', marker='o')
ax3.set_title('Gr치fico de Dispersi칩n en 3D')
ax3.set_xlabel('Eje X')
ax3.set_ylabel('Eje Y')
ax3.set_zlabel('Eje Z')

# Mostrar los gr치ficos juntos
plt.tight_layout()
plt.show()
````
